#**Bank Note Authentication**

Data were extracted from images that were taken from genuine and forged bank note like specimens. For digitalization, an industrial camera useually used for printing inspection was used.
The final images have 400 x 400 pixels. Due to the object lens and distance to the investigated object gray-scale picture with a resolution of about 660 dpi were gained.
Wavelet transformation tool were used to extract features from images.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/BankNote_Authentication.csv")
```

```python
df.head()
```

```python
 ### Independent and Dependent features

X = df.iloc[:,:-1]
y= df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score= accuracy_score(y_test, y_pred)
```

```python
score
```

```python
import pickle
pickle_out = open('classifier.pkl','wb')
pickle.dump(classifier, pickle_out)
pickle_out.close()
```
